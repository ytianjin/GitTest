```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('D:/pics/bai1.jpg')
img1 = cv.imread('D:/pics/fingjin1.jpg')
pag = cv.resize(img,(500,500))
png = cv.resize(img1,(500,500))
line = cv.addWeighted(pag,0.5,png,0.5,0)
rows,clos = line.shape[:2]
m = cv.getRotationMatrix2D((clos/2,rows/2),45,1)
line1 = cv.warpAffine(line,m,(clos,rows))
plt.imshow(line1[:,:,::-1])
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('D:/pics/bai1.jpg')
img1 = cv.imread('D:/pics/fingjin3.jpg')
img_1 = cv.resize(img,(500,500))
img1_1 = cv.resize(img1,(500,500))
pag = cv.addWeighted(img_1,0.5, img1_1,0.5,0)
rows,clos = pag.shape[:2]
png = np.float32([[50,100],[100,50],[50,200]])
png1 = np.float32([[100,50],[50,100],[100,250]])
m = cv.getAffineTransform(png,png1)
kernel = cv.warpAffine(pag,m,(clos,rows))
plt.imshow(kernel[:,:,::-1])
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('D:/pics/bai1.jpg')
img1 = cv.resize(img,(500,500))
sobel = cv.Sobel(img1,cv.CV_16S,1,0,ksize=3)
sobelx = cv.convertScaleAbs(sobel)
sobely = cv.Sobel(img1,cv.CV_16S,0,1,ksize=3)
sobely1 = cv.convertScaleAbs(sobely)
sobelxy = cv.addWeighted(sobelx,0.5, sobely1,0.5,0)
# cv.imshow('sobelxy',sobelxy)
# cv.waitKey(0)
# cv.destroyAllWindows()
plt.imshow(sobelxy[:,:,::-1])
plt.show()
```
