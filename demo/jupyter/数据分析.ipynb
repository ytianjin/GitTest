```python
import matplotlib.pyplot as plt
# %matplotlib inline

plt.figure()
plt.plot([1, 0, 9], [4, 5, 6])
plt.show()
```

```python
plt.figure()
plt.plot([1, 2, 3, 4, 5, 6, 7], [17, 17, 18, 11, 11, 15, 13])
plt.show()
```

```python
plt.figure(figsize=(20, 8), dpi=80)
plt.plot([1, 2, 3, 4, 5, 6, 7], [17, 17, 18, 11, 11, 15, 13])
# plt.savefig(保存路径)
plt.show()

# figure:创建画布, plot:绘制图像
# figsize=()画布的大小.dpi:清晰度
```

```python
import random
# from matplotlib.font_manager import FontProperties
# 1准备数据x y
x = range(60)
y_yufun = [random.uniform(15, 18) for i in x]
# 创建画布
plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y_yufun)
# 准备x的刻度说明
x_label = ["11点{}".format(i) for i in x]
# plt.xticks(range(0, 60, 5))
# plt.xticks(x[::5])
plt.xticks(x[::5], x_label[::5])
plt.yticks(range(0, 40, 5))
plt.show
```
